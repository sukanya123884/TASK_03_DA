```python
import sqlite3
import pandas as pd
conn=sqlite3.connect('mydb.db')
df_check=pd.read_sql('SELECT * FROM superstore LIMIT 5',conn)
df_check
```

### Order ID starting With CA

```python
query="""
SELECT *
FROM superstore
WHERE [Order ID] LIKE 'CA%';
"""
df_ca_orders=pd.read_sql(query,conn)
df_ca_orders
```

### Count Total Second Class Ship Mode

```python
query="""
SELECT COUNT(*) AS Second_class
FROM superstore
WHERE "Ship Mode"='Standard Class';
"""
df_second_ship=pd.read_sql(query,conn)
df_second_ship
```

### Organize the Dataset with Descending value of Profit

```python
query="""
SELECT *
FROM superstore
ORDER BY "Profit" DESC;
"""
df_profit=pd.read_sql(query,conn)
df_profit
```

### Let's see which city has the most Sales

```python
query="""
SELECT "City",SUM("Sales") as "Total Sales"
FROM superstore
GROUP BY "City"
HAVING SUM("Sales")>30000
ORDER BY "Total Sales" DESC;
"""
df_city_profit=pd.read_sql(query,conn)
df_city_profit
```

### Which Segment has bought maximum quantity of belongings ?

```python
query="""
SELECT "Segment",SUM("Quantity") as "Total Quanity"
FROM superstore
GROUP BY "Segment"
ORDER BY "Total Quantity"
LIMIT 1;
"""
df_segment_quantity=pd.read_sql(query,conn)
df_segment_quantity
```

### Top 5 Sub-Catagory with Most Sales 

```python
query="""
SELECT "Sub-Category",SUM("Sales") as "Total Sales"
FROM superstore
GROUP BY "Sub-Category"
ORDER BY "Total Sales" DESC
LIMIT 5;
"""
df_city_profit=pd.read_sql(query,conn)
df_city_profit
```
